```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import CSVLogger
import tensorflow as tf
import keras
from keras.models import Sequential
from keras import layers
from keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D, BatchNormalization

from keras.preprocessing.image import ImageDataGenerator

```

```python
## Load data
root_path = "/kaggle/input/gender-classification-dataset/Training/"
class_names = sorted(os.listdir(root_path))
n_classes = len(class_names)
class_dis = [len(os.listdir(root_path + name)) for name in class_names]
print(f"Total Number of Classes : {n_classes} \nClass Names : {class_names}")
print(f"female : {class_dis[0]} \nmale : {class_dis[1]}")
```

```python
train_gen = ImageDataGenerator(rescale=1/255.,
                               rotation_range=10, horizontal_flip=True)
valid_ds = ImageDataGenerator(rescale=1/255)
# Load Data
train_ds = train_gen.flow_from_directory("/kaggle/input/gender-classification-dataset/Training",color_mode='grayscale', 
                                         target_size=(120,90), shuffle=True, 
                                         batch_size=256) 

valid_ds = valid_ds.flow_from_directory("/kaggle/input/gender-classification-dataset/Validation", color_mode='grayscale',
                                         target_size=(120,90),
                                         shuffle=True, batch_size=256)
```

```python

model = Sequential([
    layers.Conv2D(64,(3,3),activation='relu',input_shape=(120,90,1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Dropout(0.4),
    layers.Flatten(),

    layers.Dense(128,activation='relu'),
    layers.Dense(64,activation='relu'),
    layers.Dense(2,activation='softmax')
])
```

```python
csvlogger = CSVLogger('accuracy.log',separator=',')

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit_generator(train_ds,epochs=10,validation_data=valid_ds,callbacks=[csvlogger])

# Epoch 1/10
# 184/184 [==============================] - 873s 5s/step - loss: 0.4670 - accuracy: 0.7595 - val_loss: 0.2224 - val_accuracy: 0.9141
# Epoch 2/10
# 184/184 [==============================] - 812s 4s/step - loss: 0.2311 - accuracy: 0.9092 - val_loss: 0.1620 - val_accuracy: 0.9376
# Epoch 3/10
# 184/184 [==============================] - 811s 4s/step - loss: 0.1867 - accuracy: 0.9296 - val_loss: 0.1427 - val_accuracy: 0.9466
# Epoch 4/10
# 184/184 [==============================] - 805s 4s/step - loss: 0.1733 - accuracy: 0.9350 - val_loss: 0.1307 - val_accuracy: 0.9515
# Epoch 5/10
# 184/184 [==============================] - 804s 4s/step - loss: 0.1614 - accuracy: 0.9401 - val_loss: 0.1234 - val_accuracy: 0.9561
# Epoch 6/10
# 184/184 [==============================] - 802s 4s/step - loss: 0.1527 - accuracy: 0.9446 - val_loss: 0.1243 - val_accuracy: 0.9555
# Epoch 7/10
# 184/184 [==============================] - 795s 4s/step - loss: 0.1474 - accuracy: 0.9467 - val_loss: 0.1218 - val_accuracy: 0.9579
# Epoch 8/10
# 184/184 [==============================] - 794s 4s/step - loss: 0.1412 - accuracy: 0.9492 - val_loss: 0.1163 - val_accuracy: 0.9570
# Epoch 9/10
# 184/184 [==============================] - 796s 4s/step - loss: 0.1392 - accuracy: 0.9499 - val_loss: 0.1123 - val_accuracy: 0.9593
# Epoch 10/10
# 184/184 [==============================] - 795s 4s/step - loss: 0.1331 - accuracy: 0.9531 - val_loss: 0.1186 - val_accuracy: 0.9606
# <keras.callbacks.History at 0x7d6b94ce0a60>
```
